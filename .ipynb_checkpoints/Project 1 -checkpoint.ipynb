```python
import pandas as pd 
import matplotlib.pyplot as plt 
import numpy as np 
from datetime import datetime as dt 
```

```python
#import crime.csv file 
df = pd.read_csv("../Project-1/crimedata.csv")

df.head()
df['FIRST_OCCURRENCE_DATE'].min()
```

```python
#drop some unnecessary columns 
df = df.drop(columns = ['IS_CRIME', 'IS_TRAFFIC', 'DISTRICT_ID', 'PRECINCT_ID', 'GEO_X', 'GEO_Y'])
```

```python
df.count()
```

```python
df_crime_type = df.groupby('OFFENSE_CATEGORY_ID')
df_crime_type_count = df_crime_type.count()
df_crime_type_count = df_crime_type_count.sort_values(by=['INCIDENT_ID'], ascending = False)
df_crime_type_count.head()

plt.hist(df_crime_type_count)
plt.show()
```

```python
#Only select observations within 3 years period: 2015, 2016, and 2017
df = df.loc[df['FIRST_OCCURRENCE_DATE'] > '1/1/2015']
df.count()
df = df.loc[df['FIRST_OCCURRENCE_DATE'] < '12/31/2017']
df.count()

#drop all missing values for Longitude and Latitude
df.dropna(subset = ['GEO_LON', 'GEO_LAT'])
df.count()
```

```python
#group the dataframe by offense category and find the most crime category occurance 
df_offense_category = df.groupby('OFFENSE_CATEGORY_ID')
df_offense_category_count = df_offense_category.count()
df_offense_category_count = df_offense_category_count.sort_values(by=['OFFENSE_ID'], ascending = False)
df_offense_category_count.head()

```

```python
#group the dataframe by neighborhoods and find the most crime-ridden neighborhoods  
df_neighborhood = df.groupby('NEIGHBORHOOD_ID')
df_neighborhood_count = df_neighborhood.count()
df_neighborhood_count = df_neighborhood_count.sort_values(by = ['INCIDENT_ID'], ascending = False)
df_neighborhood_count.head()
```
